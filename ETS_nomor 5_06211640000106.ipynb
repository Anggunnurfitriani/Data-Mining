```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori 
```

```python
data= pd.read_csv('online_retail.csv')
data.head()
```

```python
data.shape
```

```python
np.sum(data.isnull())
```

```python
# descriptiom memiliki beberapa komponen kosong yang harus dihilangkan
data['UnitPrice'] = data['UnitPrice'].astype('str')
data = data[~data['UnitPrice'].str.contains('0')]
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
```

```python
np.sum(data.isnull())
```

```python
invoice= data.groupby('InvoiceNo')
keranjang= []
for i, j in invoice:
    keranjang.append(list(j['Description'].map(str)))
data1=pd.DataFrame(keranjang)
```

```python
#jumlah transaksi berdasarkan nomor invoice 
len(data['InvoiceNo'].unique())
```

```python
#jumlah transaksi tanpa cust id
len(data[data['CustomerID'].isnull()]['InvoiceNo'].unique())
```

```python
#jumlah transaksi berdasarkan nomor id cust.
len(data['CustomerID'].unique())
```

```python
data1.shape
```

```python
total = []  
for i in range(0, 22465):  
    total.append([str(data1.values[i,j]) for j in range(0, 923)])
```

```python
association_rules = apriori(total, min_support=0.005, 
                            min_confidence=0.2, 
                            min_lift=3, min_length=2)  
association_results = list(association_rules)
association_results[0]
```

```python
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
