```python
#import packages, it is a must and usually people do it first
import pandas as pd #helping import a csv data
import numpy as np
```

```python
#making a table from Data Frame
left= pd.DataFrame({'key':['K0','K1','K2','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
right= pd.DataFrame({'key':['K0','K1','K2','K3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})
#when it comes to merging some data frame the first way is lookin for the same variable
result1= pd.merge(left,right, on='key')
```

```python
result1
```

```python
print(type(result1['D']))
```

```python
#making a table from Data Frame using 2 series of  data frame as criterion
left2= pd.DataFrame({'key1':['K0','K0','K1','K2'],
                   'key2':['K0','K1','K0','K1'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
right2= pd.DataFrame({'key1':['K0','K1','K1','K2'],
                    'key2':['K0','K0','K0','K0'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})
result2= pd.merge(left2,right2, on = ['key1','key2'])
result2
```
