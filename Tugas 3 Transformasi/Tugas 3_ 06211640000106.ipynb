```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv("data tugas 3.csv", sep=";")
data.head()
```

```python
proses= data.iloc[0:113,2:9]
proses.tail()
```

```python
# missing value inspection
np.sum(proses.isnull())
```

```python
proses['Speed']=proses['Speed'].fillna((proses['Speed'].mean()))
np.sum(proses.isnull())
```

```python
print(proses)
```

```python
# outlier inspection
from scipy import stats
```

```python
z= np.abs(stats.zscore(proses._get_numeric_data()))
print (z)
proses2= proses[(z<3).all(axis=1)]
# untuk mengetahui ukuran akhir data setelah menghilangkan outlier
print(proses2.shape)
```

```python
print(proses.shape)
```

```python
print(proses2)
```

```python
#normalization
from sklearn.preprocessing import normalize
prosesnorm= normalize(proses,norm='l2')
prosesnorm= pd.DataFrame(prosesnorm)
prosesnorm.tail()
```

```python
#3sigmoid nonlinear data normalization methods (1-e^(-xnormalized))/(1+ e^(-xnormalized))
proses_sigmoid= (1-2.718281828**(-prosesnorm))/(1+ 2.718281828**(-prosesnorm))
proses_sigmoid= pd.DataFrame(proses_sigmoid)
proses_sigmoid.tail()
```

```python
print(proses_sigmoid)
```

```python
proses_sigmoid.to_csv("proses3.csv")
```
